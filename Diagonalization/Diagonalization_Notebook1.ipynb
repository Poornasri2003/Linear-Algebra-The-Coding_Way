### Interpret the matrix `D` in terms of X vector. 

In linear algebra, a diagonal matrix `D` represents a linear transformation that scales each basis vector by a certain factor, while leaving the vectors themselves unchanged in direction.

In the standard basis for `R^2`, the basis vectors are usually denoted as `e_1` and `e_2`, where:

```python
e_1 = [1, 0]
e_2 = [0, 1]



So, under the transformation represented by `D`, the basis vector `e_1` is scaled by a factor of 3, and the basis vector `e_2` is scaled by a factor of 6. .


**Basis**

A basis vector is a vector that serves as one of the fundamental building blocks for a vector space. In more precise terms, a set of basis vectors forms a basis for a vector space if every vector in that space can be uniquely expressed as a linear combination of those basis vectors


````python
#importing library
import numpy as np
````

### Analyzing the usage of diagonal matrices
#### Calculating squared diagonal matrix

````python
#Declaring the diagonal matrix
D=np.array([[3,0],[0,6]])
````

````python
def calculate_square(D):
    square = D * D
    return square
````

````python
Diagonal_square = calculate_square(D)
print("Diagonal matrix")
print(D)
print("---------------")
print("\nSquared matrix")
print(Diagonal_square)
````

#### Interpretation
The diagonal matrix is very helpful in scaling . When we multiply a vector by a diagonal matrix, each element of the vector gets scaled by the corresponding diagonal element of the matrix. This operation allows us to independently scale each dimension of the vector.

#### Declaring basis e1 and e2

e1 and e2 serve as a universal basis in R2 where any other vectors can also be formed into a basis if they are linearly independent.

- e1 is  represented as 1,0
- e2 is  represented as 0,1

````python
e1=np.array([[1],[0]])
e2=np.array([[0],[1]])
````

````python
# for plotting graph
import matplotlib.pyplot as plt
````

````python
# Plotting initial e1 and e2
def plot_graph(e1,e2,labelling_x,labelling_y,title_name):
    plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='r', label=labelling_x)
    plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='b', label=labelling_y)
    
    # Set plot limits and labels
    plt.xticks(np.arange(-1, 11, 1))
    plt.yticks(np.arange(-1, 11, 1))
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title(title_name)
    # Add grid and legend
    plt.grid()
    plt.legend()
    
    # Show plot
    plt.show()
````

````python
plot_graph(e1,e2,'e1','e2','Plotting e1 and e2')
````

#### Applying transformations over diagonal matrix

#### Defining transformation function as vector multiplication

````python
def vector_multiply(D,e):
    x= np.dot(D,e)
    return x
 
````

````python
Transformation_x1=vector_multiply(D,e1)
print(Transformation_x1)
print("--------")
Transformation_x2=vector_multiply(D,e2)
print(Transformation_x2)
````

````python
plot_graph(Transformation_x1,Transformation_x2,'T(e1)','T(e2)','Plotting the Transformation "Dx"')
````

- The transformed vector does not change its direction .
- The direction remains same and e1 is scaled by 3 and e2 by 6

###  On considering different vector x  for transformation
#### Defining x vector

````python
x=np.array([[1],[1]])
````

````python
Transformation_X= vector_multiply(D,x)
print(Transformation_X)
````

````python
plot_graph(x,Transformation_X,'x','T(x)','Plotting T(x)')
````

#### Observation
The resulting vector will still point in the same direction but will be scaled differently along each axis:

 - Along the x-axis (horizontal direction), the vector will be scaled by a factor of 3.
 - Along the y-axis (vertical direction), the vector will be scaled by a factor of 6.
