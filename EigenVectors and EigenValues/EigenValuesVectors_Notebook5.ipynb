#### Importing library

```python
import numpy as np
```

#### Transition matrix
The probability matrix represents the probability of landing on a certain page when we are on a certain page. This can be termed as a transition matrix
#### Defining transition matrix

```python
transition_matrix=[[0,0,1,0.5],
                   [0.333333,0,0,0],
                   [0.333333,0.5,0,0.5],
                   [0.333333,0.5,0,0]]
```

#### Defining a Vector Transition Function

Given:
- **transition_matrix**: Probability matrix from the table.
- **X**: A vector representing the probability of landing on each page.

This function predicts the transition matrix based on:

> **transition_matrix = transition_matrix * X**

This means each element in the resulting matrix is obtained by multiplying each row of the transition matrix by the corresponding element in the vector \( X \).
or \( X \).


```python
def vector_transition(transition_matrix, X):
  for i in range(100):
    X = np.dot(transition_matrix,X)
  return X
```

**Lets say you start on Page 1, and then depending on this matrix, you move to the next page. Repeat this process, say a hundred times. How would the composition of vector x look ?**


```python
#starting at the Page1
x=[1,0,0,0] 
print("X initially when starting at page 1:", x)
# transition matrix for 100 times
for i in range(100):
    x = vector_transition(transition_matrix, x)
print("Composition of vector X after multiplying with the transition matrix for 100 times:")
#printing answers
for i, val in enumerate(x):
    print("Page {0} = {1}".format(i + 1, val))
```

**Lets say you start on Page 3, and then depending on this matrix, you move to the next page. Repeat this process, say a hundred times. How would the composition of vector x look ?**

```python
#starting at the Page3
x=[1,0,0,0] 
print("X initially when starting at page 1:", x)
# transition matrix for 100 times
for i in range(100):
    x = vector_transition(transition_matrix, x)
print("Composition of vector X after multiplying with the transition matrix for 100 times:")
#printing answers
for i, val in enumerate(x):
    print("Page {0} = {1}".format(i + 1, val))
```

**What is X**

- **X**, also referred to as the steady state vector, represents the probability distribution of landing on each page in the long run.
- When the system reaches its steady state, the probabilities in vector X remain constant over subsequent iterations, indicating that the system has reached equilibrium.
- In this context, if the probability vector **X** remains the same after 100 iterations regardless of whether the starting page is Page 1 or Page 3, it suggests that the system has converged to a stable distribution where the probabilities of landing on each page do not change significantly over time. This stable distribution is often referred to as the steady-state distribution or equilibrium distribution.
