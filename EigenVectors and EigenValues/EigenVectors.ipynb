```python

#import necessary header files
import numpy as np
import matplotlib.pyplot as plt




```

```python
# Define the matrix A
A = np.array([[1, 6],
              [5, 2]])

# Define the eigenvectors u and v
u = np.array([[6],
              [-5]])

v = np.array([[3],
              [-2]])
```

```python
# Perform matrix-vector multiplication to obtain the results Au and Av
Au = np.dot(A, u)
Av = np.dot(A, v)
```

```python
# Plotting
fig, ax = plt.subplots(1, 1, figsize=(8, 8)) # Creating a figure and axis object
ax.plot([0, u[0, 0]], [0, u[1, 0]], 'r', label='u')  # Plotting vector u
ax.plot([0, v[0, 0]], [0, v[1, 0]], 'g', label='v')  # Plotting vector v
ax.plot([0, Au[0, 0]], [0, Au[1, 0]], 'b--', label='Au')  # Plotting result of Au
ax.plot([0, Av[0, 0]], [0, Av[1, 0]], 'm--', label='Av')  # Plotting result of Av
# Setting plot limits and aspect ratio
ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_aspect('equal', 'box')
ax.grid(True)
ax.legend()
plt.title("Visualization of eigenvectors")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Compute the eigenvalues and eigenvectors of A
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigen values")
print(eigenvalues)
print("Eigen Vectors")
print(eigenvectors)
```

```python
# Check if u and v are eigenvectors of A
is_eigenvector_u = np.allclose(np.dot(A, u), eigenvalues[0] * u)
is_eigenvector_v = np.allclose(np.dot(A, v), eigenvalues[1] * v)

# Print the results
if is_eigenvector_u:
    print("u is an eigenvector of A")
    print("Eigenvalue for u:", eigenvalues[0])
else:
    print("u is not an eigenvector of A")

if is_eigenvector_v:
    print("v is an eigenvector of A")
    print("Eigenvalue for v:", eigenvalues[1])
else:
    print("v is not an eigenvector of A")
```
